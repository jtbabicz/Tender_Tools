# Calibrating XES Energy Axis from Elastic Scattering

```python
path = '/sdf/group/ssrl/dsokaras/bl6-2a/tender_data/2025-06_Tender/elastic_H2OJet_MoL3val/'
pattern = '*sif_50_080_170_0500*txt'
calibration = Tender()
calibration.load_data(path, pattern, 0)
```

```python
calibration.normalize(measurement_type = "XES", method = "Area Integration")
```

```python
calibration.combine_data(mode = "XES", incident_energies = ["2498.00eV", "2503.00eV", "2508.00eV", "2513.00eV", "2518.00eV", "2523.00eV"])
```

```python
calibration.plot("All_spectra", indices = list(range(0, 18)), fig_dim=(6, 10), label_range=[-36,-24], y_increment=0.03, bbox_values=(1.03, 1.0), title = "Water Elastic Scattering")
```

```python
calibration.plot("Final_XES", indices = [0, 1, 2, 3, 4, 5], xlims = [300, 1600], fig_dim = (6, 9), title = "Elastic Calibration", y_increment=0.003, e_calibration = None)
```

```python
calibration.fit_peaks(xlims=[600, 1550])
```

```python
calibration.fit_energy_pixel([2498.00, 2503.00, 2508.00, 2513.00, 2518.00, 2523.00])
```
